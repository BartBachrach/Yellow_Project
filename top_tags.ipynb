```python
import pandas as pd
from sklearn.cluster import KMeans, AgglomerativeClustering
import matplotlib.pyplot as plt
import hvplot.pandas
import numpy as np
```

```python
scores = pd.read_csv('/Users/adambachrach/Desktop/Yellow_Project/ml-25m/genome-scores.csv')
scores_df = pd.DataFrame(scores)
scores_df
tags = pd.read_csv('/Users/adambachrach/Desktop/Yellow_Project/ml-25m/tags.csv')
tags_df = pd.DataFrame(tags)
```

```python
scores_df[['movieId','tagId','relevance']][scores_df.relevance == scores_df['relevance'].max()]
```

```python
tags_df
```

```python
top_tags = pd.merge(scores_df, tags_df, on='movieId')
top_tags
```

```python
best_tags = top_tags.loc[top_tags.reset_index().groupby(['movieId'])['relevance'].idxmax()]
best_tags.drop(['timestamp', 'userId'], axis=1)
```

```python
best_tags2 = best_tags.drop(['tag', 'userId', 'timestamp', 'relevance'], axis=1)
best_tags2.head(50)
```

```python
clustering = AgglomerativeClustering().fit_predict(best_tags2)
```

```python
clustering
```

```python
plt.rcParams['figure.figsize'] = [150, 150]
plt.rcParams['figure.autolayout'] = True

x = np.random.randn(1500)
y = np.random.randn(1500)
centers = np.random.randn(75, 50)

fig = plt.figure()
ax = fig.add_subplot(111)

scatter = ax.scatter(x, y, c=clustering, s=50)
for i, j in centers:
    ax.scatter(i, j, s=50, c='red', marker='+')

plt.show()
```
