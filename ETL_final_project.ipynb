```python
ratings_data_to_load = "Resources/ratings.csv"
```

```python
# Read the rating data file and store it in a Pandas DataFrame.
ratings_data_df = pd.read_csv(ratings_data_to_load)
ratings_data_df
```

```python
# Determine if there are any missing values in the ratings data.
ratings_data_df.count()
```

```python
# Count ratings data
ratings_count = ratings_data_df["movieId"].count()
ratings_count
```

```python
#Determine if there are any missing ratings data
ratings_data_df.groupby('movieId',as_index=False).agg({'movieId':'count','rating':'mean'})
ratings_data_df
```

```python
# Drop unnessasary columns
ratings_data =  ratings_data_df.groupby('movieId')['movieId'].count()
ratings_data.head()
```

```python
# Obtain an average of the ratings for each movie
ratings_data = ratings_data_df.groupby('movieId',as_index=True).agg({'movieId':'count','rating':'mean'})
ratings_data.head()
```

```python
# Change name of movieId count to number of ratings
ratings_data = ratings_data.rename(columns={"movieId": "number of ratings"})
ratings_data.head()
```

```python
print (ratings_data)
```

```python
# Create CSV for average movie ratings
ratings_data.to_csv('avg_movie_ratings.csv', index = False, encoding='utf-8') 
```
